```python
import pandas as pd
import numpy as np

pd.set_option('mode.chained_assignment', None)
```

```python
df = pd.read_csv('ds_salary_2023.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df1 = df.dropna(axis=0)
```

```python
df1.shape
```

```python
df1.head()
```

```python
df1.reset_index(inplace=True)
```

```python
df1.drop('index', axis=1, inplace=True)
df1.head()
```

```python
df1.shape
```

```python
df1['year'] = df1['year'].astype('int')
```

```python
df1.head()
```

```python
pd.set_option('display.max_rows', None)
df1['salary_est'] = df1['salary_est'].apply(lambda x: x.split('(')[0])
df1['salary_est']
```

```python
df1['salary_est'] = df1['salary_est'].apply(lambda x: 0 if 'Per Hour' in x  else x.lower())
```

```python
df1.drop(df1[df1.salary_est == 0].index, axis=0, inplace= True)
```

```python
df1.shape
```

```python
df1['min_salary'] = df1['salary_est'].apply(lambda x: x.split('-')[0].replace('k', ''))
df1['min_salary'] = df1['min_salary'].apply(lambda x: int(x.replace('$', '')))
```

```python
def apply_max_salary(x):
    try:
       return x.split('-')[1]
    except IndexError:
       return 0
```

```python
df1['max_salary'] = df1['salary_est'].apply(apply_max_salary)
df1.max_salary
```
